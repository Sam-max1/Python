#### if Statement in Python

##### The 'if' statement is used for conditional execution of code.
##### It allows you to run a block of code only if a certain condition is True.

##### Syntax:
##### if condition:
#####     # Code block to execute if the condition is True

```python
# Example 1: Checking if a number is positive

number = int(input("Enter a number: "))

if number > 0:
    print("The number is positive.")
```

```python
# Example 2: Checking if a string is not empty

text = input("Enter some text: ")

if text:  # An empty string is considered False, a non-empty string is True
    print("You entered:", text)
```

```python
# Example 3: Checking if a list contains a specific element

my_list = [1, 2, 3, 4, 5]

if 3 in my_list:
    print("The list contains the number 3.")
```

```python
# Example 4: Using boolean variables

is_raining = True

if is_raining:
    print("Take an umbrella.")
```

```python
# Example 5: if statement with multiple conditions (using 'and' and 'or')

a = 10
b = 5
c = 15

if a > b and a < c:
    print("a is greater than b and less than c")

if a > 20 or b < 10:
    print("Either a is greater than 20 or b is less than 10 (or both)") #This will print because b<10 is true.
```

```python
# Example 6: if statement with 'not' operator

is_sunny = False

if not is_sunny:
    print("It is not sunny.")

# Example 7: if statement with user input

# num = int(input("Enter a number: "))
# if num > 0:
#     print("The number is positive.")
# if num < 0:
#     print("The number is negative.")
# if num == 0:
#     print("The number is zero.")

# Example 8: Checking if a list is empty
my_list = []

if not my_list:
    print("The list is empty")

my_list = [1,2,3]

if my_list:
    print("The list is not empty")

# Example 9: Checking if a string is empty
my_string = ""

if not my_string:
    print("The string is empty")

my_string = "hello"

if my_string:
    print("The string is not empty")



```

Key things to remember:

Indentation: Python uses indentation to define blocks of code. The code inside the if block must be indented.
Condition: The condition is an expression that evaluates to either True or False.
Comparison Operators: You can use comparison operators like == (equal), != (not equal), > (greater than), < (less than), >= (greater than or equal), and <= (less than or equal) to create conditions.
Logical Operators: You can use logical operators like and, or, and not to combine multiple conditions.
Truthiness: In Python, many values are considered "truthy" or "falsy." For example, non-zero numbers and non-empty strings are truthy, while zero and empty strings are falsy.


```python
# if...elif...else Statements in Python

# The if...elif...else statement is used for conditional execution with multiple branches.
# It allows you to check multiple conditions in sequence and execute different blocks of code based on which condition is true.

# Basic Syntax:
# if condition1:
#     # code to be executed if condition1 is true
# elif condition2:
#     # code to be executed if condition1 is false and condition2 is true
# elif condition3:
#     # code to be executed if condition1 and condition2 are false and condition3 is true
# ... (more elif statements as needed)
# else:
#     # code to be executed if all conditions are false

# Example 1: Simple if...elif...else

score = 85

if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
elif score >= 60:
    print("D")
else:
    print("F") #Output: B

# Example 2: Checking multiple ranges

temperature = 25

if temperature < 0:
    print("Freezing")
elif 0 <= temperature < 15:
    print("Cold")
elif 15 <= temperature < 25:
    print("Warm")
else:
    print("Hot") #Output: Hot

# Example 3: Checking for specific values

day = "Wednesday"

if day == "Monday":
    print("Start of the work week")
elif day == "Wednesday":
    print("Mid-week hump day") #Output: Mid-week hump day
elif day == "Friday":
    print("Almost weekend")
else:
    print("Just another day")

# Example 4: Using if...elif...else with user input

# num = int(input("Enter a number: "))

# if num > 0:
#     print("Positive")
# elif num < 0:
#     print("Negative")
# else:
#     print("Zero")

# Example 5: Multiple elif conditions.

value = 3

if value == 1:
    print("Value is one")
elif value == 2:
    print("Value is two")
elif value == 3:
    print("Value is three") #Output: Value is three
elif value == 4:
    print("Value is four")
else:
    print("Value is something else")

# Example 6: Checking for divisibility

number = 15

if number % 3 == 0 and number % 5 == 0:
    print("Divisible by 3 and 5") #output: Divisible by 3 and 5
elif number % 3 == 0:
    print("Divisible by 3")
elif number % 5 == 0:
    print("Divisible by 5")
else:
    print("Not divisible by 3 or 5")
```

Key points to remember:

Order Matters: The conditions are checked in the order they appear. Once a condition is true, the corresponding block of code is executed, and the rest of the elif and else blocks are skipped.
elif (else if): Allows you to check multiple conditions sequentially. You can have as many elif blocks as needed.
else (optional): The else block is executed if none of the preceding conditions are true.
Indentation: Consistent indentation is crucial in Python to define the code blocks associated with each condition.
Efficiency: Using elif is more efficient than using multiple independent if statements because it stops checking conditions once a true condition is found.
